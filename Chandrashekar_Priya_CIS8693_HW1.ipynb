```python
#Casting object types to categorical
for col in ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', 'fuel-system']:
    auto_df[col] = auto_df[col].astype('category')
```

```python
auto_df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Splitting dataset into training and test sets using 80/20 rule
y = auto_df['price']
#Dropping Price and all categorical variables
X = auto_df.drop(columns=['price', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'num-of-cylinders', 'fuel-system'])
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Running a linear regression model
model = LinearRegression()
model.fit(train_X, train_y)
```

```python
#Regression Coefficients
print('intercept ', model.intercept_)
print(pd.DataFrame({'Variables': X.columns, 'coefficient': model.coef_}))
```

```python
r2_train = model.score(train_X, train_y)
r2_train
```

```python
r2_test = model.score(test_X, test_y)
r2_test
```
