```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
flight_df = pd.read_csv("FlightDelays.csv")
```

```python
print(flight_df.shape)
```

```python
flight_df.info()
```

```python
flight_df.columns = [c.replace(' ', '_') for c in flight_df.columns]
flight_df.head()
```

```python
flight_df.dtypes
```

```python
for col in ['CARRIER', 'DEST', 'ORIGIN', 'Flight_Status']:
    flight_df[col] = flight_df[col].astype('category')
```

```python
flight_df.dtypes
```

```python
flight_df['CARRIER'] = flight_df['CARRIER'].cat.codes
flight_df['DEST'] = flight_df['DEST'].cat.codes
flight_df['ORIGIN'] = flight_df['ORIGIN'].cat.codes
```

```python
flight_df.dtypes
```

```python
flight_df.describe()
```

```python
pd.set_option("display.max_rows", None)
```

```python
flight_df.isnull().any().any()
```

```python
y = flight_df['Flight_Status']
X = flight_df.drop(columns=['Flight_Status'])
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
LogR = LogisticRegression()
LogR.fit(train_X, train_y)
```

```python
print('intercept ', LogR.intercept_[0])
print(pd.DataFrame({'coeff': LogR.coef_[0]}, index=X.columns).transpose())
```

```python
classificationSummary(train_y, LogR.predict(train_X))
```

```python
classificationSummary(test_y, LogR.predict(test_X))
```
